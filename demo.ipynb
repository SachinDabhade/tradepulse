```python
import pandas as pd
import numpy as np
from collections import deque


class RealTimeSignalGenerator:
    def __init__(self, symbol, ema_fast=21, ema_slow=55, peak_window=11):
        self.symbol = symbol
        self.ema_fast = ema_fast
        self.ema_slow = ema_slow
        self.peak_window = peak_window

        # Internal state
        self.data = pd.DataFrame(columns=['Close'])
        self.ema_fast_vals = []
        self.ema_slow_vals = []
        self.phases = []
        self.hh = []
        self.hl = []
        self.lh = []
        self.ll = []
        self.signals = []

    def update(self, new_price):
        # Append new price
        self.data.loc[len(self.data)] = new_price
        prices = self.data['Close']

        # Calculate EMAs
        ema_fast = prices.ewm(span=self.ema_fast).mean().iloc[-1]
        ema_slow = prices.ewm(span=self.ema_slow).mean().iloc[-1]
        self.ema_fast_vals.append(ema_fast)
        self.ema_slow_vals.append(ema_slow)

        # Determine Phase
        if (new_price > ema_slow) and (ema_fast > ema_slow):
            phase = "Bull"
        elif (new_price <= ema_slow) and (ema_fast < ema_slow):
            phase = "Bear"
        else:
            phase = self.phases[-1] if self.phases else "Bear"
        self.phases.append(phase)

        # Peak / Trough detection
        peak = False
        trough = False
        if len(prices) >= self.peak_window:
            window = prices[-self.peak_window:]
            peak = (prices.iloc[-1] == window.max()) and (prices.iloc[-1] != prices.iloc[-2])
            trough = (prices.iloc[-1] == window.min()) and (prices.iloc[-1] != prices.iloc[-2])

        # Trend structure
        if peak and len(prices) > 5:
            if prices.iloc[-1] > prices[-6:-1].max():
                self.hh.append(prices.iloc[-1])
            if prices.iloc[-1] < prices[-6:-1].max():
                self.lh.append(prices.iloc[-1])
        if trough and len(prices) > 5:
            if prices.iloc[-1] > prices[-6:-1].min():
                self.hl.append(prices.iloc[-1])
            if prices.iloc[-1] < prices[-6:-1].min():
                self.ll.append(prices.iloc[-1])

        # Signal logic
        signal = "HOLD"
        if phase == "Bull":
            if self.hh and new_price > self.hh[-1]:
                signal = "BUY"
        self.signals.append(signal)
        return signal
```

```python
from datetime import datetime

class MultiStockSignalManager:
    def __init__(self, symbols):
        self.symbols = symbols
        self.generators = {
            symbol: RealTimeSignalGenerator(symbol)
            for symbol in symbols
        }
        self.last_signals = {symbol: None for symbol in symbols}
        self.buy_state = {symbol: {"date": None, "triggered": False} for symbol in symbols}  # NEW

    def process_ltp(self, symbol, ltp):
        if symbol not in self.generators:
            return  # Unknown symbol

        signal = self.generators[symbol].update(ltp)

        # Handle daily BUY signal limit
        if signal == "BUY":
            today = datetime.now().date()
            state = self.buy_state[symbol]
            if state["date"] == today and not state["triggered"]:
                return  # Already sent BUY today
            # Record BUY signal
            self.buy_state[symbol] = {"date": today, "triggered": False}
        elif signal != "HOLD":
            # Optional: Reset on SELL or other new signal
            self.buy_state[symbol]["triggered"] = True

        # Avoid duplicate signal spam
        if signal != self.last_signals[symbol]:
            print(f"üîî {symbol}: {signal} at {ltp}")
            self.last_signals[symbol] = signal
            self.log_signal(symbol, signal, ltp)

    def log_signal(self, symbol, signal, price):
        with open("live_signals.csv", "a") as f:
            f.write(f"{datetime.now()},{symbol},{signal},{price}\n")

    def mark_order_executed(self, symbol):
        """Mark that the BUY order for this symbol was executed"""
        if symbol in self.buy_state:
            self.buy_state[symbol]["triggered"] = True
```

```python
import json
import os
import time
import tracemalloc
from datetime import datetime
import nest_asyncio
from dhanhq import DhanContext, MarketFeed

# Enable asyncio compatibility with notebooks or Streamlit
nest_asyncio.apply()

# Start memory tracking
tracemalloc.start()

class DhanLiveLTPStreamer:
    def __init__(self, dhan_context, security_ids, flush_interval=10, file_path="asset_current_price.json"):
        """
        Initialize the streamer.
        :param dhan_context: Authenticated DhanContext object
        :param security_ids: List of security IDs to subscribe
        :param flush_interval: How often to write to file (seconds)
        :param file_path: Path to save the LTPs as JSON
        """
        self.dhan_context = dhan_context
        self.instruments = [(MarketFeed.NSE, str(sec_id), MarketFeed.Ticker) for sec_id in security_ids]
        self.file_path = file_path
        self.flush_interval = flush_interval
        self.ltp_cache = {}  # In-memory cache
        self.feed = MarketFeed(dhan_context, self.instruments, version="v2")

    def store_in_memory(self, response):
        # Normalize response
        if isinstance(response, dict):
            responses = [response]
        elif isinstance(response, list):
            responses = response
        else:
            responses = []

        for item in responses:
            if not isinstance(item, dict) or item.get("type") != "Ticker Data":
                continue
            sec_id = str(item.get('security_id') or item.get('securityId'))
            ltp = item.get('LTP') or item.get('ltp')  # In Dhan it is 'LTP'
            if sec_id and ltp:
                self.ltp_cache[sec_id] = {
                    "ltp": float(ltp),
                    "timestamp": datetime.now().isoformat()
                }


    def flush_to_file(self):
        try:
            with open(self.file_path, "w") as f:
                json.dump(self.ltp_cache, f, indent=2)
            print(f"[{datetime.now().strftime('%H:%M:%S')}] Flushed {len(self.ltp_cache)} records to {self.file_path}")
        except Exception as e:
            print("Error writing to file:", e)

    def print_memory_usage(self):
        current, peak = tracemalloc.get_traced_memory()
        print(f"Memory usage: Current = {current / 1024:.2f} KB | Peak = {peak / 1024:.2f} KB")

    def start_streaming(self):
        print("üîÑ Starting Dhan LTP Streamer...")
        last_flush_time = time.time()

        try:
            while True:
                self.feed.run_forever()
                response = self.feed.get_data()
                print(response)
                if isinstance(response, (list, dict)):
                    self.store_in_memory(response)
                else:
                    print("‚ö†Ô∏è Unexpected response format:", type(response), response)


                if response:
                    self.store_in_memory(response)

                # Periodically write to file
                if time.time() - last_flush_time >= self.flush_interval:
                    self.flush_to_file()
                    self.print_memory_usage()
                    last_flush_time = time.time()

        except KeyboardInterrupt:
            print("‚ö†Ô∏è Streaming interrupted by user.")
        except Exception as e:
            print("‚ùå Streaming error:", e)
        finally:
            self.feed.disconnect()
            self.flush_to_file()
            print("‚úÖ Graceful shutdown completed.")
```
